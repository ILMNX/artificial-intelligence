```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from transformers import pipeline

# Initialize sentiment analysis tools
nltk_analyzer = SentimentIntensityAnalyzer()
empathetic_responses = {
    "positive": [
        "That's wonderful to hear! Keep it up!",
        "I'm glad you're feeling good. 😊"
    ],
    "neutral": [
        "I see. Would you like to share more?",
        "I'm here if you need to talk."
    ],
    "negative": [
        "I'm sorry you're feeling this way. You are not alone.",
        "It’s okay to feel like this. Can I help in any way?"
    ],
}

# GPT-based empathy generator (Optional: For more complex responses)
empathy_generator = pipeline("text-generation", model="gpt2", max_length=50)

def analyze_sentiment(user_input):
    """Analyze sentiment of user input using Vader."""
    scores = nltk_analyzer.polarity_scores(user_input)
    if scores["compound"] > 0.05:
        return "positive"
    elif scores["compound"] < -0.05:
        return "negative"
    else:
        return "neutral"

def generate_empathy(user_input):
    """Generate empathetic response using GPT."""
    prompt = f"You are a kind chatbot. Respond empathetically to: {user_input}"
    response = empathy_generator(prompt, max_length=50, num_return_sequences=1)
    return response[0]["generated_text"]

def chatbot():
    """Main chatbot loop."""
    print("Hello! I'm here to support you. How are you feeling today?")
    while True:
        user_input = input("\nYou: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Take care! I'm always here if you need support. ❤️")
            break

        # Analyze sentiment and generate a response
        sentiment = analyze_sentiment(user_input)
        print(f"Chatbot (Vader): {sentiment.capitalize()} sentiment detected.")

        # Use rule-based empathetic responses
        response = empathetic_responses[sentiment][0]
        print(f"Chatbot (Rule-based): {response}")

        # Optionally include GPT-based response
        advanced_response = generate_empathy(user_input)
        print(f"Chatbot (AI): {advanced_response}")

# Run the chatbot
if __name__ == "__main__":
    chatbot()
```
